```python
def categorize_athlate(row):
    if row['height_cm'] <175 and row['weight_kg'] <=60 :
        return 'Liight Wight'
    elif row['height_cm'] <185 and row['weight_kg']<=80:
        return 'Avrage wait'
    else:
        return 'Havy Waight'
    

bios_new['waight_cat']=bios_new.apply(categorize_athlate,axis=1)
```
