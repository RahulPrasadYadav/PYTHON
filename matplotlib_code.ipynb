```python
df=pd.read_csv('DataSet/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'] , width=.5)

```

```python
plt.bar(np.arange(df.shape[0])- 0.2, df['2015'] , width=0.2 , color='yellow' , label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='red', label='2016')
plt.bar(np.arange(df.shape[0])+ 0.2, df['2017'], width=0.2 , color='black', label='2017')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.legend()
plt.show()
```

Stacked BAr chat 

```python
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2017'], label='2016')
plt.bar(df['batsman'], df['2015'], bottom=(df['2016']+ df['2017']), label='2015')
plt.legend()
plt.show()
```

```python
iris=pd.read_csv('New folder/iris.csv')

iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-versicolor':1 , 'Iris-virginica':2, 'Iris-setosa':0})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'] , c=iris['Species'], cmap='winter', alpha=0.5)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')

plt.colorbar()
```
