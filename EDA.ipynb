```python
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd plot
sns.kdeplot(df['Seles(unit)'], shade=True, color='green')
plt.title('Density Plot of Monthly Sales', fontsize=16)
plt.xlabel('Sales (Units)', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.tight_layout()
plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Sales data for each month
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June', 
              'July', 'August', 'September', 'October', 'November', 'December'],
    'Sales (Units)': [100, 120, 150, 130, 110, 170, 160, 180, 200, 190, 220, 250]
}
df = pd.DataFrame(data)
# Create a figure with two subplots (1 row, 2 columns)
plt.figure(figsize=(18, 6))

# Bar Chart
plt.subplot(1, 2, 1)
sns.barplot(x='Month', y='Sales (Units)', data=df, palette='Blues_d')
plt.title('Monthly Sales Bar Chart', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie Chart
plt.subplot(1, 2, 2)
plt.pie(df['Sales (Units)'], labels=df['Month'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Blues_d', len(df)))
plt.title('Sales Distribution by Month', fontsize=16)

```

# Calculate 3-month moving average


```python
df['Moving Average (3 Months)'] = df['Sales (Units)'].rolling(window=3).mean()
```

```python
plt.plot(df['Month'], df['Sales (Units)'], marker='o', label='Monthly Sales', color='b', linestyle='-', markersize=8)
plt.plot(df['Month'], df['Moving Average (3 Months)'], label='3-Month Moving Average', color='r', linestyle='--', linewidth=2)

plt.legend()
```


Numerical vs. Categorical Data:
Box Plot:
Box Plot: Displays the distribution of the numerical variable for each category. It helps to visualize how the numerical data is spread within each group


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'C'],
    'Values': [5, 7, 8, 6, 10, 9, 11]
})

# Boxplot
sns.boxplot(x='Category', y='Values', data=df)
plt.title('Boxplot of Values by Category')
plt.show()
```

Categorical vs. Categorical Data:
When both variables are categorical, common methods include:
Contingency Table (Crosstab): A table that shows the frequency of occurrences for each combination of categorical variables.
Chi-Square Test: A statistical test used to determine whether there is a significant association between two categorical variables.
Stacked Bar Plot: A visual method for showing how the distribution of one categorical variable change across different categories of the other.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'A']
})

# Crosstab (Contingency Table)
crosstab = pd.crosstab(df['Gender'], df['Product'])
print(crosstab)

# Stacked Bar Plot
crosstab.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot of Gender vs Product')
plt.ylabel('Count')
plt.show()

```

<!-- Multivariate Analysis -->
Multivariate Analysis is a statistical technique used to analyse data that involves more than two variables at the same time. It helps researchers and analysts understand relationships, patterns, and dependencies among multiple variables, allowing them to study the combined effects of these variables and make more informed decisions.
Objectives of Multivariate Analysis:
To examine the relationships between multiple variables.
To identify patterns and structures within the data.
To make predictions based on multiple input variables.
To reduce the dimensionality of large datasets while retaining important information.
Plots for Multivariate Analysis
Pair Plot (for visualizing pairwise relationships between variables)
Principal Component Analysis (PCA) (for dimensionality reduction to 2D)
Heatmap (for correlation matrix)
3D Scatter Plot (for visualizing 3 features)

Pairplot: It is also known as a scatterplot matrix, is a type of visualization used to understand the relationships between multiple variables in a dataset.

<!-- Key Features of a Pair Plot: -->

Diagonals

Symmetry
<!-- Purpose of Pair Plot -->

```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Species'] = data.target_names[data.target]

# Create the pair plot
sns.pairplot(df, hue='Species')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply PCA to reduce data to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df.iloc[:, :-1])

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Species'] = df['Species']
```

```python
import seaborn as sns
# Correlation matrix
corr_matrix = df.iloc[:, :-1].corr()
# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap - Iris Dataset')
plt.show()
```
