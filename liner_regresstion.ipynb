# Modal Implemention 
# simple Liner Regression (Study Time )

```python
import numpy as np 
import  matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score
```

```python
study_time=np.array([1,2,3,4,5,6,7,8]).reshape(-1, 1)

test_scores=np.array([50,55,65,70,75,80,85,88])
```

```python
study_time
test_scores
```

# spliting the dataset into  trainin and testing Sets  we 
#  use 80% of the data for Trainin  and 20% for Testing 


```python
X_train, X_test, y_train , y_test=train_test_split(study_time, test_scores , test_size=0.2, random_state=0)
```

```python
X_train
```

```python
X_test
```

# Training the Liner reggrestion Modal 
# Create an instence of the LIner Regresstion class and fir thw modal


```python
modal= LinearRegression()
modal.fit(X_train, y_train)
```

```python
# Step 5: Visualizing the Regression Line for Training Data
# Plot the training data points
model = LinearRegression()
model.fit(X_train, y_train)
plt.scatter(X_train, y_train, color='blue', label='Training Data')


# Plot the regression line using the training data
plt.plot(X_train, model.predict(X_train), color='red', label='Regression Line')
plt.title('Training Data: Study Time vs Test Scores')
plt.xlabel('Study Time (hours)')
plt.ylabel('Test Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 6: Making Predictions on Test Data
# Predict the test scores for the test set
y_pred = model.predict(X_test)
```
